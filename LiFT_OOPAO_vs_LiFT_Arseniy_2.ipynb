```python
D = 10.0 # [m]
resolution = 200 # telescope resolution (pupil sampling)

ang_pixel_Nyquist_rad = 0.5*(1.654e-06)/D

pixel_size = 18e-6 # [m]
ang_pixel = 50 # [mas]
print(ang_pixel)
f = pixel_size / ang_pixel * 206264806.71915 # [m]
print(f)

# Let's imagine we have DITs (detector integration times) of 10 frames with a total of 1 second exposure
# This is to generate a synthetic PSF 
sampling_time = 0.1 # [s]
num_samples = 10

# If the number of pixels in image is odd, the code will automatically center generated PSFs it to one pixel in the middle
tel_A = Telescope_A(img_resolution        = 9,  # img_resolution - Sampling of the telescope's PSF
                    pupil             = mask_circle(resolution, resolution/2, center=(0,0), centered=True),
                    diameter          = D,
                    focalLength       = f,
                    pupilReflectivity = 1.0,
                    gpu_flag          = False)


print("Telescope's pupil area = {} [m^2]".format(tel_A.area))

det_A = Detector_A(pixel_size     = pixel_size,
                sampling_time = sampling_time,
                samples       = num_samples,
                RON           = 0.0, # used to generate PSF or the synthetic R_n [photons]
                QE            = 0.7) # not used

det_A.object = None
det_A * tel_A
ngs_A = Source_A([('H', 14.0)]) # Initialize a target of H_mag=10
ngs_A * tel_A # attach the source to the telescope
```

```python
# Source and Zernike base for OOPAO
src = Source(optBand = 'H', magnitude = 14)
src*tel

##%% -----------------------     ZERNIKE    ----------------------------------
nModes = 50               # number of zernike to compute

#create a zernike object
Z = Zernike(telObject = tel, J = nModes)

# compute the zernike polynomials associated to the telescope tel
Z.computeZernike(tel)
```

```python
det_pixel_size = 18e-6
det_ang_pixel = 50 # mas
f = det_pixel_size / det_ang_pixel * 206264806.71915 # [m]

det = Detector_Rafael(tel=tel, pixel_size=det_pixel_size, ang_pixel=det_ang_pixel, img_resolution = 9, samples=1, RON=0, QE=1)

diversity_shift = 200e-9 #[m]

amp_diversity = np.zeros(nModes)
amp_diversity[3] = diversity_shift

OPD_diversity_OOPAO = np.squeeze(Z.modesFullRes@amp_diversity)

def PSFfromCoefs(coefs,OPD_diversity_OOPAO):

    OPD = np.squeeze(Z.modesFullRes@coefs)
    tel.OPD = OPD + OPD_diversity_OOPAO
    k = 2*np.pi/(tel.src.wavelength)
    amplitude = np.sqrt(tel.pupilReflectivity * tel.pupil/tel.pupil.sum() * tel.src.nPhoton * ((np.pi*(tel.D)**2)/4) * tel.samplingTime)
    # plt.imshow(amplitude)
    # plt.colorbar()
    # plt.show()
    # print(tel.pupil.sum())
    PSF = PropagateField(tel, det, np.sqrt(tel.src.fluxMap), k*tel.OPD, tel.src.wavelength, return_intensity=True, oversampling=1)

    # PSF = det.PropagateField(np.sqrt(tel.src.fluxMap), k*tel.OPD, tel.src.wavelength, return_intensity=True, oversampling=1)

    # print(np.shape(PSF))
    # PSF = PropagateField(tel, det, amplitude, k*tel.OPD, tel.src.wavelength, return_intensity=True, oversampling=1)
    
    # tel.OPD *= 0
    # tel.src.phase *= 0

    return PSF

def PSFfromCoefs_A(coefs):
    tel_A.src.OPD = Z_basis_A.wavefrontFromModes(tel_A,coefs) + OPD_diversity
    PSF = tel_A.ComputePSF()
    tel_A.src.OPD *= 0.0 # zero out just in case
    return PSF

# Initialize modal basis
Z_basis_A = Zernike_A(modes_num = 50)
Z_basis_A.computeZernike(tel_A)

OPD_diversity = Z_basis_A.Mode(3)*diversity_shift

coefs_0 = np.zeros(50,dtype=float)
coefs_0[:10] = np.random.uniform(-50, 50, 10)*1e-9 # [m]  random coefficients

make_3_plot_graph(PSFfromCoefs(coefs_0,OPD_diversity_OOPAO),"OOPAO PSF",PSFfromCoefs_A(coefs_0),"PSF Arseniy",\
                  PSFfromCoefs(coefs_0,OPD_diversity_OOPAO)-PSFfromCoefs_A(coefs_0),"Difference")

print(tel_A.oversampling)
```

## MAP (OOPAO) vs MAP (Arseniy)
- comment the checkbackend in Arseniy function before running this: this way, we can use "model"
- when using MAP we can never achieve the very low errors we achieve with ML at least when we have no noise
- using MAP with mean zero and var=1 is the same as ML
- MAP may be useful only for realistic conditions with noise and high-order residuals
