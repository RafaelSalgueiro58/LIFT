# Tests with LiFT for real data on sky from Keck

```python
from astropy.io import fits
import os
import numpy as np
import matplotlib.pyplot as plt

from modules.Telescope import Telescope
from modules.Detector  import Detector
from modules.Source    import Source
from modules.Zernike   import Zernike
from modules.LIFT      import LIFT

from tools.misc import mask_circle
from tools.fit_gaussian import fitgaussian, gaussian
```

```python
crop = 24

ang_pixel     = 10.0 # mas
sampling_time = 1
pixel_size    = 27e-6 # [m]

band = "K"

keck = Telescope(img_resolution    = crop,
                 pupil             = mask_circle(64, 32, center=(0,0), centered=True), # a circular pupil is assumed
                 diameter          = 11.732,
                 focalLength       = pixel_size / ang_pixel * 206264806.71915,
                 pupilReflectivity = 1.0,
                 gpu_flag          = True)

det = Detector(pixel_size    = pixel_size,
               sampling_time = sampling_time,
               samples       = 1,
               RON           = 38.0,
               QE            = 0.7)

ngs = Source([(band, 0.0)])
# ngs = Source([(band, 8.0)])
det * keck
ngs * keck

# Initialize modal basis
Z = Zernike(modes_num=50)
Z.computeZernike(keck)

diversity_shift = 200e-9    
new_diversity = Z.Mode(4)*diversity_shift

estimator = LIFT(keck, Z, new_diversity, 20)
```

```python
# Define the relative path to the folder containing the data
data_folder = 'Keck_data_Carlos/20181120'

# Define the absolute path to the notebook directory
notebook_dir = os.getcwd()  # This gets the current working directory of the notebook

# Construct the absolute path to the data folder
data_path = os.path.join(notebook_dir, data_folder)

# List all files in the data folder
files = os.listdir(data_path)

print(files)

defocus_values_est = []

# Load each FITS file
for index,file_name in enumerate(files[:-1]):
    if file_name.endswith('.fits'):
        file_path = os.path.join(data_path, file_name)
        print(file_path)
        # Load the FITS file
        hdul = fits.open(file_path, mode='readonly', ignore_missing_end=True)
        # hdul = fits.open(file_path)

        # Access the first HDU (indexing starts at 0)
        hdu = hdul[0]
        
        # Extract the data from the HDU
        data = hdu.data
        print("Shape of data images: {}".format(np.shape(data)))
        
        data = data.astype("float")
        data -= np.median(data)

        first_crop = data[400-5:440-5,600-10:630]

        plt.imshow(first_crop)
        plt.colorbar()
        plt.show()

        # Find the index of the maximum value in the flattened array
        max_index_flat = np.argmax(first_crop)
        
        # Convert the flat index to 2D coordinates
        max_index_2d = np.unravel_index(max_index_flat, first_crop.shape)
        
        print("Coordinates of maximum value:", max_index_2d)

        # second_crop = first_crop[max_index_2d[0]-12:max_index_2d[0]+12,max_index_2d[1]-12:max_index_2d[1]+12]
        second_crop = first_crop[max_index_2d[0]-12:max_index_2d[0]+12,max_index_2d[1]-12:max_index_2d[1]+12]

        plt.imshow(second_crop)
        plt.colorbar()
        plt.show()

        PSF_0 = second_crop/np.sum(second_crop)

        plt.imshow(PSF_0)
        plt.colorbar()
        plt.show()

        # Perform phase retrieval with LiFT

        modes = [0,1,2,3,4,5]

        A_1_init = np.zeros(len(modes))

        A_1, _, _ = estimator.Reconstruct(PSF_0, R_n="model", mode_ids=modes, A_0=A_1_init, optimize_norm='sum')

        print(A_1)

        print("Defocus estimated: {}".format(A_1[2]))

        defocus_values_est.append(A_1[2])
        
        plt.plot(A_1,".")
        plt.ylabel("Estimated amplitudes (rad rms)")
        plt.xlabel("Zernike mode")
        plt.grid()
        plt.show()
        
        # Optionally, access the header information
        header = hdu.header
        # print(header)
        
        # Close the FITS file
        hdul.close()
```

```python
plt.plot(np.arange(1,len(files)),np.array(defocus_values_est)*10**9,"o")
plt.grid()
plt.show()
```
